# Simulation of the Conway's Game of Life

A minimalist notebook that let you simulate a game of life providing the world's dimension and the number of turns

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
import IPython.display

dimension = int(input('What is the world dimension ?'))
turns = int(input('What is the number of turns ?'))
```

```python
world = np.random.randint(2, size=(dimension,dimension))

def world_to_fig(world, cmap='viridis'):
  global img
  fig = plt.figure()
  img = plt.imshow(world, cmap=cmap, interpolation="nearest", vmin=0)
  plt.title('World')
  plt.close()
  return fig

def sum_nb(row, col):
    global world
    nb = []
    # We iterate over neighbour rows
    for nb_row in (-1, 0, +1):
        for nb_col in (-1, 0, +1): 
            # we exclude the target cell
            if nb_row!=0 or nb_col!=0:
                nb.append(world[(row + nb_row) % dimension,(col + nb_col) % dimension])
    
    return sum(nb)

def next_turn(turn):   
    global world
    sum_of_nb = np.zeros((dimension, dimension))

    # We compute a new matrix containing the sum of neighbours for each cell
    for row in range(dimension): 
        for col in range(dimension): 
            sum_of_nb[row, col] = sum_nb(row,col)

    # We make the world evolve according to the GoL rule
    world = (world & (sum_of_nb == 2)) | (sum_of_nb == 3)
    img.set_array(world)
    return img,

    
fig = world_to_fig(world, cmap='binary')
IPython.display.HTML(anim.FuncAnimation(fig, next_turn, frames=50, interval=50).to_jshtml())
    
```
